## Q1 - Manipulando o DataFrame

Crie uma coluna chamada `Hemisfério` baseada na Latitude

A regra de formação é a seguinte:

Valor | Critério
---|---
Norte | Latitude positiva
Sul | Latitude negativa

```python
C = []
V = []
for i in df.Latitude:
    if i > 0:
        C.append("Latitude positiva")
        V.append("Norte")
    if i < 0:
        C.append("Latitude negativo")
        V.append("Sul")
```

```python
df["Hemisfério"] = pd.Series(V)

```

## Q6 - Tirando conclusões com base nos dados

Em um certo lugar já ocorreram abalos com *Magnitude Type* `MB` e *Type* `Nuclear Explosion`. 
 
Responda: 
    * É mais provável que tenha sido no norte ou no sul? 
    
Assuma que os Magnitude Type e Type são independentes


```python
#TOTAL DE POSSIBILIDADES
x = df[(df.Type == "Nuclear Explosion") & (df['Magnitude Type'] =="MB")]
n = x.shape[0]
#TOTAL DE OCORRENCIAS NO NORTE
x1 = df[(df.Type == "Nuclear Explosion") & (df['Magnitude Type'] =="MB") & (df.Hemisfério == "Norte")] 
y1 = x1.shape[0]
#TOTAL DE OCORRENCIAS NO SUL
x2 = df[(df.Type == "Nuclear Explosion") & (df['Magnitude Type'] =="MB") & (df.Hemisfério == "Sul")] 
y2 = x2.shape[0]
```

```python
#PORCENTAGEM DE TER OCORRIDO NO SUL:
print(y2*100/n)
```

```python
#PORCENTAGEM DE TER OCORRIDO NO NORTE:
print(y1*100/n)
```
